```python
# Import libraries and define the alias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Discard warnings of future changes in package
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Import datasets
netflix = pd.read_csv('datasets_tp/netflix_dataset.csv')
countries = pd.read_excel('datasets_tp/countries.xlsx')
weather = pd.read_csv('datasets_tp/austin_weather.csv')
weather2013 = weather[weather['Year']==2013]
weather2014 = weather[weather['Year']==2014]
weather
```

#### First Plots <a class="anchor" id="P1.1"></a>

- Plot y versus x as lines and/or markers.

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html <br>
*Axes.plot(args, scalex=True, scaley=True, data=None, kwargs)*

```python
fig, ax = plt.subplots()
x = np.arange(-3, 3, 0.01)
def normal(x):
    return (2*3.14159)**(-1/2) * np.exp(-(x**2))

ax.plot(x, normal(x))
```

```python
# Single Plot
fig, ax = plt.subplots() # conventioned interface
ax.plot(weather2014['Date'], weather2014['TempLowF']) # (x,y)
plt.show() # required to display the plot
```

```python
#how you could fix the axis
import matplotlib.dates as mdates
fig, ax = plt.subplots() # conventioned interface
ax.plot(weather2014['Date'], weather2014['TempLowF']) # (x,y)
ax.xaxis.set_major_locator(mdates.DayLocator(bymonthday=1))  # Ticks on the 1st of each month
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b')) 
plt.xticks(rotation=45)
plt.show() # required to display the plot
```

```python
# Mutiple Plot
fig, ax = plt.subplots()
ax.plot(weather2014['Date'], weather2014['TempLowF']) # Daily lowest temperature
ax.plot(weather2014['Date'], weather2014['TempHighF']) # Daily highest temperature
plt.show()
```

Bar plot <br>
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html

```python
# This bar chart expects the "height" for y value (it isn't a histogram!)
plt.style.use('default') #restore default style
fig, ax = plt.subplots(figsize=(13,4))
ax.bar(countries.loc[:25,'country'], countries.loc[:25,'pop2023']) # (x, height)
ax.set_xticklabels(countries.loc[:25,'country'], rotation=45)
plt.show()
```

```python
# Stacked version
c_list = countries.loc[:25,'country'] # x value - countries list
h1 = countries.loc[:25,'pop2023'] # height 1
h2 = countries.loc[:25,'pop2030'] # height 2
h3 = countries.loc[:25,'pop2050'] # height 3
# Note: it doesn´t make sense to stack this values but it's just for demonstrative purposes

fig, ax = plt.subplots(figsize=(13,4))
ax.bar(c_list, h1)
ax.bar(c_list, h2, bottom=h1) # bottom defines what's under this bar (pop2023): RELEVANT FOR NEXT WEEK!!!!
ax.bar(c_list, h3, bottom=h1+h2) # bottom defines what's under this bar (pop2023 and pop2030)
ax.set_xticklabels(c_list, rotation=45)
plt.show()
```

Share axis <br>
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.sharex.html <br>
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.sharey.html
<br><br>[Back to TOC](#toc)

```python
# Share x-axis range
fig, ax = plt.subplots(2,3, figsize=(14,10), sharex=True) # x ticks label only appears in bottom plots
ax[0,0].plot(weather2013['Date'], weather2013['TempAvgF'])
ax[0,1].plot(weather2013['Date'], weather2013['DewPointAvgF'])
ax[0,2].plot(weather2013['Date'], weather2013['HumidityAvgPercent'])
ax[1,0].plot(weather2013['Date'], weather2013['SeaLevelPressureAvgInches'])
ax[1,1].plot(weather2013['Date'], weather2013['VisibilityAvgMiles'])
ax[1,2].plot(weather2013['Date'], weather2013['WindAvgMPH'])
plt.show()
```

```python
# Share y-axis range
fig, ax = plt.subplots(1,2, figsize=(14,4), sharey=True) # y ticks label only appears in left plot
ax[0].plot(weather2013['Date'], weather2013['VisibilityAvgMiles'])
ax[1].plot(weather2013['Date'], weather2013['WindAvgMPH'])
plt.show()
```

Relplot <br>

- This function allows to create subplots in a single figure and can set its kind (scatter or line plot).

https://seaborn.pydata.org/generated/seaborn.relplot.html
<br><br>[Back to TOC](#toc)

```python
# Replicate the line plot
sns.relplot(x=weather2014['Date'], y=weather2014['TempLowF'], kind='line') # Generalized line plot
plt.show()
```

```python
# Same, in a scatter plot...
sns.relplot(x=weather2014['Date'], y=weather2014['TempLowF'], kind='scatter') # The default type
plt.show()
```

```python
#...but, to be even better, it should be defined as:
sns.relplot(data=weather2014, x='Date', y='TempLowF', kind='scatter') # set 'data' and only use columns' name for other args
plt.show()
```

```python
# Example of subplot (with just the first trimester)...
sns.relplot(data=weather2014[:90], x='Date', y='TempLowF', kind='scatter', col='Month')
plt.show()
```

```python
# ...to plot all the months and avoid a single col or row, can define the 'col_wrap'
sns.relplot(data=weather2014, x='Date', y='TempLowF', kind='scatter', col='Month', col_wrap=3)
plt.show()
```

Using 'hue' to color the points

```python
# Color the points based on the months
sns.relplot(data=weather2014, x='Date', y='TempLowF', hue='Month') # Hue is usually a categorical variable
plt.show()
```
