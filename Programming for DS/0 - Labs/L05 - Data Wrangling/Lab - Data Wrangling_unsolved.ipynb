```python
# Import libraries and define the alias
import pandas as pd
import numpy as np
```

```python
# Import dataset
sleep = pd.read_csv('datasets_lab/sleep_DF.csv')
sleep
```

 **Exercise 0:** Reshape the dataset with .melt()

```python
# Melt the sleep dataset by 'Person ID' and label the variables as 'features' and the values as 'record'; 
# and record it as 'df_melted1'

df_melted1 = sleep.melt(
    id_vars="Person ID",
    value_vars = sleep.drop(columns="Person ID").columns,
    var_name = "features",
    value_name = "record"
)

df_melted1
```

```python
# Melt the sleep dataset by 'Person ID', 'Gender' and 'Age', and record it as 'df_melted2'
df_melted2 = sleep.melt(
    id_vars = ["Person ID", "Gender", "Age"]
)

df_melted2
```

**Exercise 1:** Reshape the dataset with .wide_to_long()

```python
# Get 'Sleep Duration' and 'Sleep Disorder' in one column (stubnames argument), along with 'Person ID' (i argument)
pd.wide_to_long(
    sleep,
    i = "Person ID",
    stubnames = "Sleep ",
    j = "SleepInfo",
    suffix='\\w+'
)
```

```python
# As in previous exercise, use 'Person ID', 'Gender' and 'Age' in the index
pd.wide_to_long(
    sleep, 
    i = ["Person ID", "Gender", "Age"],
    stubnames = "Sleep ",
    j = "Type",
    suffix='\\w+'
)
```

 **Exercise 2:** Reshape the dataset with .stack()

```python
# Stack the dataset with default arguments and record it as 'df_true'
df_true = sleep.stack()
df_true
```

```python
# Stack the dataset with the argument dropna=False and record it as 'df_false'
df_false = sleep.stack(dropna=False)
df_false
```

```python
# Using the previous 2 recorded variables, how many null values exist in 'sleep'? 
df_false.shape[0] - df_true.shape[0]
```

 **Exercise 3:** Reshape the dataset with .unstack()

```python
# Unstack 'df_true'
df_true.unstack()
```

```python
# Unstack 'df_false'
df_false.unstack()
```

 **Exercise 4:** Reshape the dataset with .pivot()

```python
# Pivot 'df_melted1' to get the original dataset (12 columns)
pd.pivot(
    df_melted1,
    index = "Person ID",
    columns = "features"
)
```

```python
# Pivot 'df_melted2' to get the original dataset (12 columns)
pd.pivot(
    df_melted2,
    index = ["Person ID", "Gender", "Age"],
    columns = "variable"
).reset_index(level=2)
```

 **Exercise 5:** Reshape the dataset with .pivot_table()

```python
# Pivot 'df_melted1' to get the original dataset
df_melted1.pivot_table(
    index = "Person ID",
    columns = "features",
    aggfunc = "sum"
)
```

 **Exercise 6:** Reshape the dataset with .pivot_table()

```python
sleep.dtypes
```

```python
# For each numeric column of 'sleep', get its average value by 'BMI Category' (index) and 'Occupation' (columns)
phi = (sleep.select_dtypes(exclude="O").merge(sleep[["Person ID","BMI Category", "Occupation"]], on="Person ID")).pivot_table(
    index = "BMI Category",
    columns = "Occupation",
    aggfunc = "mean",
)

for x in sleep.select_dtypes(exclude="O").columns:
    display(phi[x])
```

 **Exercise 7:** Reshape the dataset with .transpose()

```python
# Without swapping index and columns variables, transpose the previous output
# For each numeric column of 'sleep', get its average value by 'BMI Category' (index) and 'Occupation' (columns)
(sleep.select_dtypes(exclude="O").merge(sleep[["Person ID","BMI Category", "Occupation"]], on="Person ID")).pivot_table(
    index = "BMI Category",
    columns = "Occupation",
    aggfunc = "mean",
).T
```
