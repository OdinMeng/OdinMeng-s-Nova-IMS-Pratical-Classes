# Smanettamento con matplotlib / seaborn boh...

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate synthetic data (random walk)
pos = 0 # Starting point p0 = 0
data = [pos] # Track pos

k = 10*(10**5)
for iter in range(k-1):
    choice = np.random.normal()
    pos += choice
    data.append(pos)
```

```python
df = pd.DataFrame(pd.Series(data))
df.columns = ["Position"]
```

```python
# Plot with numpy
fig, ax = plt.subplots()
sns.set_style("white")
ax.plot(df.index, df["Position"])
plt.axhline(df["Position"].mean(), color="r", linestyle="--")
```

```python
fig, ax = plt.subplots()
ax.hist(df["Position"])
```

```python
# Generate synthetic categorical data
genders = ["male", "female", "unicorn"]
data_2 = []

k = 100
for iter in range(k-1):
    choice = np.random.choice(genders)
    data_2.append(choice)
```

```python
# Visualize synthetic categorical data
df_2 = pd.DataFrame(data_2, columns=["Gender"])
df_2.head()
```

```python
df_2i = df_2.copy()
df_2i["dummy"]=1
sorted_columns = df_2i.groupby("Gender").sum().sort_values(by="dummy", ascending=False).index.tolist()
```

```python
# SNS + PLT
sns.set_style('darkgrid')
fig, ax = plt.subplots()

plt.suptitle("Gender of Planet Earth")

plt.sca(ax)
my_palette = {"male":"#0099CC", "female":"#FF0066", "unicorn":"#FFFF00"}
sns.countplot(x="Gender", hue="Gender", data=df_2, palette=my_palette)
plt.xlabel("Gender", rotation=45)
plt.title("yo", style="italic", fontweight="bold")
plt.ylabel("Frequency", rotation=100)
```
