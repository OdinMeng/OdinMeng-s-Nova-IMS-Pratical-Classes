# Convergenza dominata di Lebesgue

**Lemma.** Sia $(f_n)_n:\mathbb N \mapsto \mathbb R \longrightarrow \mathbb R$ la successione di funzioni a valor reali definita come
$$
f_n(x)=e^{-nx^2}
$$
Allora il limite dell'integrale di Lebesgue ha il seguente valor:
$$
\lim_n \int_\mathbb R f_n = \int_\mathbb R \lim_n f_n = 0
$$

**Proof.** Si nota subito che $f_n$ ha la funzione dominante $\tilde f$ definita come
$$
\tilde f(x) := e^{-x^2} \stackrel{\forall n}\geq f_n(x)
$$
Siccome $\tilde f \in L^1$, abbiamo che applicando il *teorema della convergenza dominata di Lebesgue* segue la tesi. $\blacksquare$

Adesso facciamo una visualizzazione di questo risultato

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib as mpl

def f_n(n, x):
    return np.exp(-n*(x**2))

X = np.arange(-3, 3, 0.01)
Y = f_n(1, X)

fig, ax = plt.subplots()

sns.lineplot(x=X, y=Y, label=r"$\tilde f(x)=e^{-x^2}$", color="r")


k = 12

# Generate colors list
p = sns.color_palette("crest", n_colors = k)

for i in range(2, k):
    sns.lineplot(x=X, y=f_n(i*2.5, X), color = p[i-2],size=1/(i-1), ax=ax, legend=False)
```

Questo  un caso tipico in cui si usa il *teorema della convergenza dominata di Lebesgue*, ove effettuiamo lo scambio tra l'integrale di Lebesgue effettuatosi nello spazio $(\Omega, \Sigma, \mu)$ e il limite:
$$
\lim_n \int_\Omega f_n\  \text d \mu= \int_\Omega (\lim_n f_n)\  \text d \mu
$$

Tuttavia ci richiede la presenza della funzione dominante $\tilde f \in L^1$. Infatti, ci sono casi in cui ci non vale.

**Controesempio.** Sia $f_n$ la successione delle funzioni uniformi, definite come
$$
f_n(x) = \frac{\chi_{[0, n]}(x)}{n}
$$
Notiamo preliminarmente che una sua propriet fondamentale  
$$
\forall n \in \mathbb N, \int_\mathbb R f_n = \frac{n}{n}=1=:\lVert f_n \rVert_{L^1}
$$
Adesso calcoliamo la sua funzione *target* di convergenza. Siccome $f_n \in O(n^{-1})$, abbiamo che il suo limite  banalmente $\lim_n f_n(x) = 0$.

Tentiamo di applicare il teorema della convergenza dominata di Lebesgue.
$$
\lim_n \int_\mathbb R f_n \stackrel{?}= \int_\mathbb R (\lim_n f_n) = \int_\mathbb R 0 = 0
$$
Tuttavia si ha la contraddizione con la sua propriet fondamentale, ovvero $\lVert f_n \rVert_{L^1} = 1, \forall n \in \mathbb N$. Da ci concludiamo che non si pu sempre applicare il *teorema della convergenza dominata di Lebesgue*, specie quando non si  in grado di trovare la funzione dominante $\tilde f \in L^1$.

$$
\exists (f_n)_n \in L^1: \lim_n \int_\mathbb R f_n \neq \int_\mathbb R \lim_n f_n
$$

```python
def uniform(n, x):
    return (x>=0)*(x<=(n))*(1/n)


k = 8

X = np.arange(0, 10, 0.005)


fig2, ax2 = plt.subplots()


# Generate colors list
p = sns.color_palette("crest", n_colors = k)
for j in range(1, k):
    i = j
    sns.lineplot(x=X, y=uniform(i, X), color = p[j-1], size = 1/i, ax=ax2, legend=False)
```
