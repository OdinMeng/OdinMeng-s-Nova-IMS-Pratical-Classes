```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Creating Figures and Axes

```python
m,n = (1,1)
x,y = (2, 2)

fig, axes = plt.subplots(
    nrows = m, ncols = n, # Specify how many axes to put in our figure
    figsize = (x, y), # Specify the size of our figure (in inches)
    sharex = False, sharey = False, # Specify whether to share the x/y-line over the axes
) # fig is a Figure object, axes could be either an Ax object or a collection of Ax objects
```

### Customizing Figures and Axes

```python
fig, axes = plt.subplots(
    nrows = 1, ncols = 5,
    figsize = (15, 2), 
    sharex = False, sharey = False,
) # Rewrite for replicability
ax = axes[0]

# Customize text on ax
ax.set_xlabel("hello") # Set string label on x axis
ax.set_ylabel("world") # Set string label on y axis

ax.set_title("foo") # Set title of the axis

ax.set_xticks([1, 2]) # Change structure of the ticks (x-axis)
ax.set_yticks([10, 20]) # Analogous as above

ax.set_xticklabels(["good", "bad"]) # Change the label of the ticks
ax.set_yticklabels(["start", "end"])

ax.spines['top'].set_visible(False) # Modify spines appearance
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

ax.grid(True) # Decide whether to let grid show or not

# Customize text on fig
fig.suptitle("bar") # Title of figure (may overwrite on axis)

# Export figure
fig.savefig("./my_graph.png", dpi=250, transparent=True)

ax.legend()
```

### Basic Plotting with Matplotlib

```python
fig, ax = plt.subplots(
    nrows = 1, ncols = 1,
    figsize = (15, 2), 
    sharex = False, sharey = False,
)
# -------------------------------
# Synthetic data (sigmoid function)
x = np.arange(-10, 10, 1)
y = (1)/(1+np.exp(-x))

# Lineplots and its customization (usually same with others)
ax.plot(
    x, y, # Define the x and y values of the plot; each datapoint will be collected with a line (or nothing if the dx is small enough)
    marker = "X", # Set marker type
    linestyle = ":", # Set line type
    color = "r", # Can be either the initial, its 3-uple code (R,G,B) or its HEX code
    linewidth = 1, # Specify sizes
    markersize = 4,
)
```

```python
# Bar plots and its customization
fig, axes = plt.subplots(
    nrows = 2, ncols = 1,
    figsize = (10, 5), 
    sharex = False, sharey = False,
)
# -------------------------------
# Synthetic data (Categories with random distributions)
cat = ["A", "B", "C"]
cat_freq = [1, 5, 3]
cat_freq_stack = [2,3,10]

# Synthetic data (Exponential function)
x = np.arange(0, 10, 1)
y = np.exp(x)

# Histogram from categorical data
ax1 = axes[0]
ax1.bar(
    cat, cat_freq, # x and y are the categories and its associated numerical value
    edgecolor = "b"
)
ax1.bar(
    cat, cat_freq_stack, 
    bottom = cat_freq, # Stack on top of something (bottom)
    color = "#FFFF00", # Specify color fill
    edgecolor = "r" # Specify color of edge (default: none)
) # If you want to stack more in an increasing manner, you have to sum the previous "stacks"

# Histogram from numerical data (discretization)
ax2 = axes[1]
ax2.bar(
    x, y,
    width = 1 # Spcify width of binss
)
```

```python
# Boxplots
fig, ax = plt.subplots(
    nrows = 1, ncols = 1,
    figsize = (4,4), 
    sharex = False, sharey = False,
)

# -------------------------------
# Synthetic data (Gauss numbers)
x = [np.random.normal(0, 1) for i in range(20)]

ax.boxplot(x)
```

```python
# Scatterplots
fig, ax = plt.subplots(
    nrows = 1, ncols = 1,
    figsize = (4,4), 
    sharex = False, sharey = False,
)

# -------------------------------
# Synthetic data (Gauss numbers in 2D)
k = 1000
x1 = [np.random.normal(0, 2) for i in range(k)]
y1 = [np.random.normal(0, 2) for i in range(k)]

x2 = [np.random.normal(2, 1) for i in range(k)]
y2 = [np.random.normal(4, 0.5) for i in range(k)]


ax.scatter(x1,y1, color="r", alpha=.5)
ax.scatter(x2,y2, color="b", alpha=.5) # Obs. You can stack multiple plots into one axis
```

# Plotting with Seaborn

```python
# Setting style or color palettes in seaborn

sns.set_style("dark") # Set to one of their predefined styles
sns.color_palette(
    "crest",
    n_colors = 12 # Specify number of colors in the palette
) # Returns an "iterable" palette item
```

```python
# Line and scatter plots
fig, axes = plt.subplots(
    nrows = 1, ncols = 2,
    figsize = (10, 5), 
    sharex = False, sharey = False,
)

# Generate synthetic data (parabola and its number scale)
x = np.arange(-100, 100, 10)
y = x**2
z = []
for i in range(len(x)):
    z.append(np.random.choice(["A","B","C"]))


my_palette = sns.color_palette(
    "hls",
    3
)

sns.scatterplot(
    x = x, y= y,   
    hue = z, # Specify hue, which categorizes each datapoint
    palette = my_palette, # Specify palette for hue
    ax = axes[0] # Specify which ax to work on
)

sns.lineplot(
    x = x, y = y,
    ax = axes[1]
)

# sns.relplot generalizes both of them, though the only work figure-level
```

```python
# Matrix plots

# Generate synthetic matrix
M = []
for i in range(1, a:=5-1):
    v_i = []
    for j in range(1, b:=5-1):
        v_i.append(np.random.normal(0, 1))
    M.append(v_i)

# Generate heatmap
my_palette = sns.color_palette(
    "vlag",
    as_cmap = True
)
sns.heatmap(M, cmap=my_palette)
```
